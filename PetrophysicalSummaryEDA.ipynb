```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file= 'https://github.com/orkahub/PEG_Python/raw/master/Data/Sample%20Petrophysics.xlsx'
```

```python
df = pd.read_excel(file)
```

```python
df.plot('Porosity', 'Sw', kind='scatter')
```

```python
zones = df['Zones'].unique()
zones
```

```python
df[df['Zones']=='MONTREAL']
```

```python
for zone in zones:
    df[df['Zones']==zone].plot('Porosity', 'Sw', kind='scatter')
```

## Using Seaborn library

```python
for zone in zones:
    sns.boxplot(df[df['Zones']==zone]['NTG'])
    plt.title(zone)
    plt.show()
```

```python
for zone in zones:
    sns.distplot(df[df['Zones']==zone]['NTG'])
    plt.title(zone)
    plt.show()
```

```python
sns.pairplot(df, hue='Zones')
```
